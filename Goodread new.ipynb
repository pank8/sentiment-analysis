```python
#раскрыть комментарии
def read_more (buttons):
    for each_button in buttons:
        try:
            each_button.click()
        except:
            pass
```

```python
#след страница
def next_pages (arrows):
    for each_arrow in arrows:
        try:
            each_arrow.click()
        except:
            pass
        
```

```python
import time

path = r'C:/Users/sesa514967/Downloads/chromedriver_win32/chromedriver.exe'

if path != r'C:/Users/sesa514967/Downloads/chromedriver_win32/chromedriver.exe':
    print('Check your path')
    
else:
    #перейти на сайт
    driver = webdriver.Chrome(path)
    driver.maximize_window()
    driver.implicitly_wait(10)
    driver.get(f'https://www.goodreads.com/choiceawards/best-{term}-books-2018/')

    #окна
    window_before = driver.window_handles[0]
    window_before_title = driver.title
    window_after = driver.window_handles[-1]

    #перейти по ссылке на книгу
    link_to_winner = driver.find_element_by_xpath("//*[@class= 'winningTitle choice gcaBookTitle gcaBookTitle--red']").get_attribute('href')
    driver.switch_to.window(window_after)
    driver.get(link_to_winner)

    #кнопка закрыть фейсбук
    try:
        close = driver.find_element_by_css_selector("body > div:nth-child(5) > div > div > div.modal__close > button")
        close.click()
    except:
        pass

    results = {}

    record_number = 0
    #while driver.find_element_by_class_name("next_page").get_attribute('href'):
    
    buttons = driver.find_elements_by_link_text("...more")
    for each_button in buttons:
        try:
            each_button.click()
        except:
            pass
    
    for counter in range(5):


        reviews = driver.find_element_by_css_selector("#bookReviews")
        N = len(reviews.find_elements_by_class_name('review'))

        for i in range(N):
            review = reviews.find_elements_by_class_name("review")[i]
            review_id = review.get_attribute('id')
            name = driver.find_element_by_css_selector(f"#{review_id} > div.left.bodycol > div.reviewHeader.uitext.stacked > span > a").text
            date = driver.find_element_by_css_selector(f"#{review_id} > div.left.bodycol > div.reviewHeader.uitext.stacked > a").text
            comment = driver.find_element_by_css_selector(f"#{review_id} > div.left.bodycol > div.reviewText.stacked").text
            try:
                rating = driver.find_element_by_css_selector(f"#{review_id} > div.left.bodycol > div.reviewHeader.uitext.stacked > span.staticStars.notranslate").text
            except:
                rating = "-"
            results[record_number] = {"Review ID":review_id,"Name":name,"Date":date, "Comment":comment, "Rating":rating}
            record_number += 1



        arrow = driver.find_element_by_link_text("next »")
        webdriver.ActionChains(driver).move_to_element(arrow).click(arrow).perform()
        #arrow.click()
        time.sleep(2)
```

```python
results_pd = pd.DataFrame().from_dict(results).transpose()
results_pd['Comment'] = results_pd['Comment'].apply(lambda s: s.replace('\n',' '))
pd.set_option('display.max_colwidth',150)
```
